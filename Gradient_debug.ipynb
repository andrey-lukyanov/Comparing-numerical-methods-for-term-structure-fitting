```python
from new_thesis_library import * 
import matplotlib.pyplot as plt
```

# Without gradient

```python
theta_0 = np.array([4.836731, 0.08779514, -0.03113246, 0.00511053])

theta, time = optimize_on_day_with_starting_values(date_number = 0, method = 'L-BFGS-B', theta0 = theta_0)

m = np.linspace(0.1, 30, 60)


plt.plot(m,  rep_ns(m, theta_0), label = 'Initial')
plt.plot(m,  rep_ns(m, theta), label = 'Optimization')
plt.legend()
plt.grid();
print('Theta: ', theta, time, 'sec.')
```

# With gradient

```python
theta_0 = np.array([4.836731, 0.08779514, -0.03113246, 0.00511053])

theta, time = optimize_on_day_with_starting_values(date_number = 0, method = 'L-BFGS-B', theta0 = theta_0)

m = np.linspace(0.1, 30, 60)


plt.plot(m,  rep_ns(m, theta_0), label = 'Initial')
plt.plot(m,  rep_ns(m, theta), label = 'Optimization')
plt.legend()
plt.grid();
print('Theta: ', theta, time, 'sec.')
```
